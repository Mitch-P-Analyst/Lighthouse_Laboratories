## Seaborn Assignment

```python
# import pandas
import pandas as pd

# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv',sep=',')
df.drop('#',axis=1,inplace=True)
```

```python
df.head()
```

### Task 1: Plot a regression line between attack and defense.

```python
# Set plot size and style (optional)
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Plot regression line
sns.regplot(x='Attack', y='Defense', data=df)

# Titles and labels
plt.title('Regression Line: Attack vs Defense')
plt.xlabel('Attack')
plt.ylabel('Defense')

# Show plot
plt.show()
```

### Task 2: Plot a regression line between attack and defense while displaying a distinction between the different generations. 

```python
# Set plot size and style (optional)
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Plot regression line
sns.lmplot(x='Attack', y='Defense', data=df, hue='Generation')

# Titles and labels
plt.title('Regression Line: Attack vs Defense')
plt.xlabel('Attack')
plt.ylabel('Defense')

# Show plot
plt.show()
```

### Task 3: Make a boxplot for each feature of dataframe (without Generation and Legendary features) in one plot.

### Task 4: Plot a correlation heatmap of the pokemon dataframe (not including the features Generation and Legendary).

```python
pm_spec = df.drop(columns=['Generation', 'Legendary'])

pm_spec
```

```python
# correlation matrix
correlationmatrix = pm_spec.corr(numeric_only=True)
# f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(correlationmatrix, vmax=.8, square=True,cmap="RdYlGn_r")
```
