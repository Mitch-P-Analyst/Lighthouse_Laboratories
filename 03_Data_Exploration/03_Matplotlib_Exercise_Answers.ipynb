## Matplotlib Assignment

```python
# import pandas
import pandas as pd

# import matplotlib
import matplotlib.pyplot as plt
```

```python
# This datasource doesn't allow us to access it directly 
# We need to modify read_csv's user agent 
# Basically, we're accessing as if through a Firefox Browser

url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'
storage_options = {'User-Agent': 'Mozilla/5.0'}
df = pd.read_csv(url, storage_options=storage_options)
```

```python
df
```

### Task 1: Read the `total_profit` of all months and display it using a line plot.

```python
total_profit = df['total_profit']
months = df['month_number']

fig = plt.figure()

monthly_profit = fig.add_axes([0.1,0.2,0.8,0.9])

monthly_profit.plot(months,total_profit,'green')
monthly_profit.set_xlabel('months')
monthly_profit.set_ylabel('profit')
monthly_profit.set_title('monthly profits')
```

### Task 2: Read all the different product sales data and display it using a multiline plot.

```python
total_profit = df['total_profit']
months = df['month_number']

fig = plt.figure()

monthly_profit = fig.add_axes([0.1,0.2,0.9,1.1])

# monthly_profit.plot(months,total_profit,'green')
monthly_profit.set_xlabel('months')
monthly_profit.set_ylabel('profit')
monthly_profit.set_title('monthly profits')

months = df['month_number']
facecream = df['facecream']
facewash = df['facewash']
shampoo = df['shampoo']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
moisturizer = df['moisturizer']

ax = monthly_profit
ax.set_ylim([0, 16000])

ax.plot(months,facecream, label='Facecream', color='purple')
ax.plot(months,shampoo, label='Shampoo', color='blue')
ax.plot(months,toothpaste, label='Toothpaste', color='yellow')
ax.plot(months,bathingsoap, label='Bathingsop', color='lightblue')
ax.plot(months,moisturizer, label='Moisturizer', color='green')
ax.plot(months,facewash, label='Facewash', color='violet')


ax.legend()
```
